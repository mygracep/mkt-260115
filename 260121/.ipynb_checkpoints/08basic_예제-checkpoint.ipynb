```python
# price 라는 변수에 있는 100 달러의 값을 환율을 적용해서 한화로 변환해서 값을 출력 
price = "100 달러"
exchange_rate = 1480

 
won = int(price) * int(exchange_rate)
won()
```

```python
# 해설 
# 1. 문자열을 숫자로 치환 : 슬라이싱으로 100 찾아옴 -> int로 숫자로 형변환
print(int(price[0:3]))
won = int(price[0:3]) * int(exchange_rate)
print(f"{won:,}원")  # :, => 천 단위마다 , 표기 
```

```python
반복문 활용해서 구구단 2단~9단까지 출력 

- 2~9까지 숫자가 필요 
- 변수 1개를 1~9까지 순서대로 곱하기
- 그걸 7번 반복해야 함 
- 그걸 {변수} x 1 = {값} 의 형태로 출력해야함 
    
```

```python
# 반복문 활용해서 구구단 2단~9단까지 출력 
for index_1 in range(2,10) : 
    print(f"====구구단 {index_1}단====")
    for index_2 in range(1,10) :
        print(f"{index_1} X {index_2} = {index_1}* {index_2}")
    print("\n") # \ = escape sequence => 문자처럼 생겼지만, 문자 이상의 고유한 문법을 가지고 있는 문자 
```
