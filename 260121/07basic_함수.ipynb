# 파이썬 함수 (funcion = 기능)

- 함수 : 어떤 입력값을 받아서 내부에서 연산처리 후 해당 결과값을 외부로 출력할 수 있도록 하는 것
- input & output : 함수에 값 입력 & 출력
- 함수 생성 = 함수를  선언한다  -> 함수 존재 -> 값 입력( 값 = 인자값 / 인수 = 매개변수) 
- 함수 사용 = 함수를 호출한다
- 
- 반복적인 실행코드를 자동화하고 싶을 때
- *함수의 선언부 & 호출부의 "인자값/인수=매개변수"의 갯수는 같아야함
   (한쪽이 상대적으로 갯수가 많거나 적으면 실행 x)
  
- return : 함수 내부연산 처리 후 특정 값을 함수 바깥으로 값을 보낼 수 있는 기능

```python
def test_func1(a, b) :
    quot = a // b  # // = 몫
    rem = a % b  
    return(quot , rem)

result = test_func1(10, 3)


# 구조분해 할당 식 => 뒤에 있는 값을 갚에 넣어줌
(f, s) = result # 등호 기준으로 위치 바뀌면 안됨
print(f, s)
```
