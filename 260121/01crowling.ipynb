# 크롤링
-웹사이트에서 내가 원하는 부분을 자동으로 추출해오는 기능 
1) requests : 정적인 웹사이트(html, css, js)
2) selenium : spa 방식으로 구현 (rect.js, vue,ss, angluar,js)
3) scrapy : 여러 절차를 거쳐서 방대한 데이터를 가져와야 하는 경우 


```python
import requests # reqeests => 데이터를 수집해오는 크롤러 
import bs4  # bs4 => 찾아온 데이터를 쉽게 이해하고 해석할 수 있는 형태로 바꿔주는 역할 
url = "https://news.daum.net/tech"
headers = {
    "Uuser-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/144.0.0.0 Safari/537.36 Edg/144.0.0.0"
}
respons = requests.get(url, headers=headers)
respons.encoding = "utf-8" #  "utf-8" => 한국어를 컴퓨터가 이해할 수 있는 언어로 변환 
soup = bs4.BeautifulSoup(respons. text, "html.parser") # "html.parser" =>html 방식으로 읽어라 

articels = soup.select ("ul.list_newsheadline2 li a.item_newsheadline2")  
# select : 다수의 값 찾아와서 리스트 형태로 반환 / selext_one : 최초의 값 1개만 찾아옴 
# 포함되는 테그 띄어쓰기 그 아래 테그 띄어쓰기 그 안에 클래스 기입 

for article in articels :
    title = article.text.strip()
    link = article["href"]
    print(f"{title} \n {link}")

```

```python
#위키피디아에서 '마케팅' 검색 후 결과값 크롤링 

import requests
import bs4

url = "https://ko.wikipedia.org/wiki/%EB%A7%88%EC%BC%80%ED%8C%85"
headers = {
    "User-Agent" : "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/144.0.0.0 Safari/537.36 Edg/144.0.0.0"
}
rp = requests.get(url, headers=headers)
respons.encoding = "utf-8"
soup = bs4.BeautifulSoup(rp.text, "html.parser")

pg = soup.select("div.mw-content-ltr.mw-parser-output > p") # p가 들어가는 이유?? - 찾아와야할 대상이 p라서 

for p in pg[:2] :
    print(p.get_text().strip())
```

```python
#문제 실습

import requests
import bs4

url = "https://www.weather.go.kr/w/forecast/overall/short-term.do"
headers = {
    "User-Agent" : "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/144.0.0.0 Safari/537.36 Edg/144.0.0.0"
}
rp = requests.get(url, headers=headers)
respons.encoding = "utf-8"
soup = bs4.BeautifulSoup(rp.text, "html.parser")

respons = soup.select("p.summary") 
# summary : 여러번 쓰였을수 있기 때문에 동일 단어가 없는지 찾아와야함. 만약 동일한 단어가 있으면 포함하고 있는 상위 테그를 찾아와야함  

for r in respons : 
    print(r.get_text("\n"))
```

```python
# 답변 
import requests
import bs4

url = "https://www.weather.go.kr/w/forecast/overall/short-term.do"
headers = {
    "User-Agent" : "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/144.0.0.0 Safari/537.36 Edg/144.0.0.0"
}
rp = requests.get(url, headers=headers)
respons.encoding = "utf-8"
soup = bs4.BeautifulSoup(rp.text, "html.parser")

tag = soup.select_one("p.summary") 
# summary : 여러번 쓰였을수 있기 때문에 동일 단어가 없는지 찾아와야함. 만약 동일한 단어가 있으면 포함하고 있는 상위 테그를 찾아와야함  

clean_text = tag.get_text("\n")
print(clean_text)

```
