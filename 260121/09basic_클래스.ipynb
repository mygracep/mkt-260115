# 프로그래밍 언어 역사
- 절차지향 언어(PP) : c ,파스칼 코볼
> 절차를 다 지켜서 진행
> 장정: 직관, 명확
> 단점: 매번 절차를 지켜서 반복되는것도 다 적어야함 
ex) 냉장고 문을 연다 - 코끼리를 냉장고에 넣는다 - 냉장고 문을 닫는다 x  반복 


- 객체지향언어(OOP)  Object(객체) Oriented Programming : JS, Java, C++, C#, python
> 객체 : 변경할 수 있는 것
ex) 코끼리라는 대상의 정보를 메뉴얼(클래스)로 생성
> - 해당 메뉴얼 기반으로 코끼리1(객체) 생성
> - 냉장고라는 대상의 정보를 메뉴얼(클래스)로 생성
> - 해당 메뉴얼을 기반으로 냉장고1(객체)를 생성
> - 코끼리1 객체가 가지고 있는 기능을 사용해 냉장고1 객체의 문을 열게 함
> - 코끼리1 객체가 가지고 있는 기능을 사용해 냉장고1 객제 안으로 들어가게 한다
> - 코끼리1 객체가 가지고 있는 기능을 사용해 냉장고1 객체의 문을 닫게 한다 
>- => 객체만 변경하면 됨

```python
# 생성자 함수 : 초기 클래스가 작동할 수 있게 세팅해주는 함수 
class Student :
    def __init__ (self, name, grade) :
        self.name = name
        self.grade = grade 

    def study(self) : 
        print(self.name,"공부합니다")
    def exam(self, subject) : 
        print(self.name, subject, "공부합니다")
```

```python
student1 = Student("David", 3)
student2 = Student("jane", 3)

print(student1.name, student1.grade)
print(student2.name, student2.grade)
```

```python
student1.study()
```

```python
student1.exam("수학") 
```
