```python
import requests
from bs4 import BeautifulSoup

res = requests.get("https://crawlingtest.netlify.app")
soup = BeautifulSoup(res.text, "html.parser")



# section = soup.select("dev_course_list")
# print(section)

section =soup.find("ul", id="dev_course_list") 
titles = section.find_all("li", "course")

#dnumerate = 셈하다 => 넘버링

for index, title in enumerate(titles) :
    print(index+1, "댓글 수:", title.get_text(" ", strip=True).split("-")[-1].replace("\n", " ").replace(" ","")[-2:-1])
    # [-2:-1] => 맨 끝에서: 그 앞에꺼까까지 찾아오기 = 슬라이싱 

```

```python
import requests
from bs4 import BeautifulSoup

res = requests.get("https://crawlingtest.netlify.app")
soup = BeautifulSoup(res.text, "html.parser")

items = soup.select("ul > li")
for item in items : 
    print(item.get_text(" ", strip=True).replace("\n","").replace(" ",""))
```
