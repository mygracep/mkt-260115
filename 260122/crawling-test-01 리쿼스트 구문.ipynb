# requests 구문 

- select: (선택자 기준으로 찾아옴) 복수의 값을 찾아올때, 유사 리스트 형태로 값을 반환으로 "반복문 사용 가능"
- select_one : 단수의 값을 찾아올때 
- text : 크롤링 해온 값에서 "문자만 남기고" 싶을 때 사용
- strip() : 해당 문자 앞, 뒤 빈 여백 혹은 이스케이프 시퀀스를 제거하고자 할때 
-  . = 클래스  # = 아이디 
- ("속성명") : ["속성값"]
- get_text(" ", strip=True)
- find : (테그 기준으로 찾아옴) 단수 값 찾아옴
- find_all : 복수의 값을 찾아옴 

find 수식 
> card = soup.find("div", class_="product-card")  테그, calss_= 값
> 생략형 data = soup.find("p", "cssstlye")
> 속성값 data = soup.find("p", attrs={"allign": "center"}) attrs(속성약자)={"속성": "속성값"}
> 아이디값 data = soup.find(id="body")

크롤링 할때 값 찾기 추천 
1. select / select_one
2. find & find_all (중복 테그가 없을시 사용 가능)

```python
url ="https://crawlingtest-260122.netlify.app/"
# headers 값은 딕셔너리로 넣어야 함 
headers = {
    "Uuser-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/144.0.0.0 Safari/537.36 Edg/144.0.0.0"
}
import requests 
from bs4 import BeautifulSoup 
respons = requests.get(url, headers=headers)
respons.encoding = "utf-8" 
soup = BeautifulSoup (respons.text, "html.parser")


items = []

for card in soup.select(".product-card") :
   # price = int(card.select_one{".price"}["data-price"])
    items.append({
        "name": card.select_one(".product-name").text.strip(),
        "category": card.select_one(".category").text.strip(),
        "price": int(card.select_one(".price")["data-price"])
    })
    
print(items)
    
```
