```python

url ="https://crawlingtest-260122.netlify.app/"
# headers 값은 딕셔너리로 넣어야 함 
headers = {
    "Uuser-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/144.0.0.0 Safari/537.36 Edg/144.0.0.0"
}

import requests 
from bs4 import BeautifulSoup 

respons = requests.get(url,headers=headers)
respons.encoding = "utf-8" 

soup = respons.text, ("html.parser")

print(response.text)

```

```python
# 상품 이름 찾아오기 예제 
name = soup.select("div.product-list h2")


print(name)
```

```python
# 상품명 여러개 존재 
# 공통 태그 proudct-name
# 한번에 3개 찾아와야 함 
# 1개의 변수에 3개의 값을 포함해야함 
# 리스트 형태로 가져와야 함 

for tag in soup.select(".product-list") :
    print(tag. text.strip()) 
# strip => 문자열 객체에만 사용 가능 / 문자로 바꿔주어야만 사용 가능 

#리스드로 변경 => fltmxm zjavmflgostus
names = [tag.text.strip() for tag in soup.select(".product-list")]
print(names)
```

```python
# 상품명. 카테고리 문제

name = soup.select("div.product-card")

for n in name :
    pn = n.text.strip()
    c = n.select("category")
    print(f"{c}-{pn}")
```

```python
# 해설 

# card = soup.select("div.product-card") => 별도의 수식 쓰지 않고 바로 for문에 넣음 

for card in soup.select(".product-card") :
    pn = card.select_one(".product-name").text.strip()
    c = card.select_one(".category").text.strip()
    print(f"{c}-{pn}")
```

```python

- select: 복수의 값을 찾아올때, 유사 리스트 형태로 값을 반환으로 "반복문 사용 가능"
- select_one : 단수의 값을 찾아올때 
- text : 크롤링 해온 값에서 "문자만 남기고" 싶을 때 사용
- strip() : 해당 문자 앞, 뒤 빈 여백 혹은 이스케이프 시퀀스를 제거하고자 할때 
-  . = 클래스  # = 아이디 
- ("속성명") : ["속성값"]
- get_text(" ", strip=True)
```
