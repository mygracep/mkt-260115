```python
import pandas as pd

input_csv = "seoul_news_p1_to_p5_with_dept_date.csv"
df = pd.read_csv(input_csv) # df = data frame 

df["modified_date"] = pd.to_datetime(df["modified_date"], errors="coerce") 
df["modified_date"] 
df = df.dropna(subset=["modified_date"])

df["month"] = df["modified_date"].dt.to_period("M").dt.to_timestamp()
df["week"] = df["modified_date"].dt.to_period("W").dt.to_timestamp()
df["dept_group"] = df["dept"].str.split("-").str[0].str.strip()

df
```

```python
import plotly.express as px

share = df["dept_group"].value_counts().reset_index()
fig = px.pie(share, names="dept_group", values="count", title="컨텐츠 점유율")
fig.show()
```

```python
dept_counts = df["dept"].value_counts().head(10).reset_index()

fig = px.bar(dept_counts, x="dept", y="count", title = "부서별 컨텐츠 작성 수 TOP 10")
fig.update_layout(xaxis_tickangle=-45)
fig.show()
```
