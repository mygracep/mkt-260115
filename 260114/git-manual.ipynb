# git 사용법
1) 로컬 컴퓨터 (내 컴퓨터) : git 응용 프로그램 다운 및 설치
2) 터미널 : git -v 명령어
   - 검색바 > cmd > 창 클릭 : 검정색 창
3) 만약, 없다면 > 구글 크롬 > git 다운로드 후 설치
   - chocolatey : 대신 설치해주는 곳
4) git의 소유주 확인 & 변경
   - git config --list
   - user email : mygracep@gmail.com
   - user name : mygracep
   - 터미널 재접속 > 관리사모드 : 전역(global) 세팅 -모든곳
   - git config --global user.name "mygracep"
   - git config --global user.email "mygracep@gmail.com"

5) github 회원가입 : 계정정보 값으로 입력
   - git : 나의 로컬컴퓨터에서 버전 관리 툴
   - github : 나의 로컬컴퓨터에서 관리하고 있는 버전을 타인과 공유해서 함께 사용하고자 할 때, 혹은 내가 사용하고 있는 디바이스가 2대 이상일 때 (A,B 기기 모두에서 데이터를 관리하고 싶을 때)
   - repository : 프로젝트별 응대

6) git commit & push
   - commt: 라벨붙이기 push: 깃허브에 넘기기  
   -  git init : 한 프로젝트에 1번만 사용 가능 중복 사용하면 충돌됨 100메가 제한
   -  c드라이브> 사용자>user> dataclass_weekday 폴더 찾아서 주소 복사
   -  cd C:\Users\user\dataclass_weekday  입력-> 현재 나의 루트폴더로 이동
   -  git 초기화 : git init -> .git 빈폴더 생성됨
   -  staging area : 데이터 전송  git add .
   -  대이터 tracking : git staus
   -  데이터 labeling : git commit -m "230115"
   -  원격 (앱) github : git remote add origin https://github.com/mygracep/mkt-260115.git -> 레포지토리 주소
   -  로컬(git) & 원격(github) 연결 확인: git remote -v
   -  로컬 -> 원격 데이터 푸쉬 : git push origin master
