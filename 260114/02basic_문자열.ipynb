   - str() : 문자열 변환

   - list() : 리스트 자료구조 변환

   - tuple() : 튜플 자료구조 변환
   - set() : 세트(집합) 자료구조 변환
     
- 문자열 : 독립적인 함수 혹은 속성들이 많이 존재
   - count() : 특정 문자열에서 어떤 단어가 몇 번 사용되었는지를 집계하고자 할 때 (대, 소문자 구분)
   - len() : length = 길이 (총 몇글자인지)
   
  순서/위치 잦기-> 문자열 안에 각각의 문자는 고유한 숫자를 가지고 있음  항상 0부터 시작
     *수식: str.find/index(sub, start, end)
            sub: 찾고자 하는 문자열 (필수), start: 검색을 시작할 위치 (선택, 기본값 0), end: 검색을 끝낼 위치 (문자열 끝)
            => 가장 처음에 나온 위치 값만 표기 
   - find() : 문자열만 해당 / 없을때 -1 표기 
   - index() : 리스트, 튜플, 문자열 같은 자료형에서 특정 데이터가 몇 번째 위치에 있는지 찾아줌. * 값이 없을 때 에러(ValueError) 발생
      * find / index -> 가장 처음에 나온 위치만 추출
   - indexing : 변수[] (변수= 문자열/리스트, 딕셔너리) 찾고 싶은 위치에 있는 데이터 반환
                index()는 어느 위치에 있는지 반환/ indexing은 해당 위치에 있는 데이터 반환 

   - replace() : 특정 문자열을 대체하고자 할때 
   - slicing : 문자열/리스트 에서 특정 구간을 잘라내서 가져오고 싶을 때

   - strip() : 공백제거. / text.strip(",") => 괄호 안에 값을 넣으면 해당 값 삭제
     
   - formet() : 특정 문자열을 원하는 위치에 출력하고자 할때 사용할 수 있는 메서드(method) 함수
               => f-string = f"{변수} {변수}" 

- 형식지정자 : 특정 타입의 값을 출력하고자 할때 어떤 형태로 값을 찾아와서 출력할 것인가 결정

```python
#find

text = "banana"

index = text.find("a")
next_index = text.find("a", index + 1) # 1번째, 그 다음번째( 2번째) 인덱스 찾기 
next_next_index = text.find("a", next_index + 1)  # 3=3번째 위치한 문자 /  +1= 그 다음번째 인덱스
print(index, next_index, next_next_index)
```

```python
#find 반복문 사용 

text = "banana"
target = "a"

idx = text.find(target) # 존재하지 않는 문자를 찾을 경우 = -1 값 

# 예외조항 처리 : 어떤 명령문을 실행시키고자 할때, 만약 ~가 아니라면, 해라!! 

while idx != -1 : 
    print(idx)
    idx = text.find(target, idx + 1)
```

```python
#find 반복문 예제

news = "오늘 뉴스입니다. 오늘 날씨는 맑음입니다. 오늘 하루 즐겁게 보내세요."
search_word = "오늘"
index = news.find(search_word)

positions = [] # 위치를 담을 바구니

while index != -1:
    positions.append(index) # 찾은 위치(0, 11, 23...)를 기록
    index = news.find(search_word, index + 1) # 다음 "오늘"을 찾으러 출발!

print(f"'{search_word}'은(는) 총 {len(positions)}번 발견되었고, 위치는 {positions}입니다.")
```

```python
text = "python"
print(text) # 문자 통채로 찾아오기 
print(text[5]) # indexing [] ~번째 인덱스 값을 찾아옴  
# 어던 문자열이던, 마지막번째 문자의 인덱스 넘버는 전체 문자열 개수 -1 

print(len(text)) # 문자 개수
print(len(text)-1) # 전체 -1 개수
print(text[len(text)-1]) # 전체 개수 -1 의 문자 출력
print(text[-1]) # 오늘쪽, 가장 끝에 문자를 찾는건 -1부터 시작 
```

```python
#slicing

text = "python"
print(text)
print(text[2])

print(text[2:4]) # 슬라이싱: 찾아오고자 하는 문자열 [start:finish+1]
print(text[2:]) # 슬라이싱에서 마지말번째 인덱스의 생략 = 끝까지 
print(text[2:len(text)]) #->해당 구문 많이 사용 
#=> 마지막까지 찾아올 수 있는건 이 3가지 문구 

print(text[2:-1]) #마지막 바로 전 문자까지 찾기 가능 / -는 뒤에서부터 시작 / -1은 뒤에서 첫번째 
```

```python
#farmet 
# 객체 - 메서드 함수 
# 형식: 문자열 형식의 객체(반드시 필요). formet()

name = "I have a {}, I have an {}."
print(name)

print(name.format("pen", "apple")) # 포멧 괄호 인자값은 순차적으로 중괄호에 들어감 
print("I have a {1}, I have an {0}".format("pen", "apple")) # 포멧 괄호 인자값은 순차적으로 중괄호에 들어감 
```

```python
interest = 0.0891
print(format(interest, ".2f")) # 독립함수 (문자, 숫자 x) -> 소수점 위치 일치 

#파이썬 내장함수 가운데 소수점을 올리거나 내리는 독립 기능 x
#외부 라이브러리 모듈 찾아서 사용 
#수학갹체 라이브러리 모듈 

import math 
result01 = math.floor(interest) # floor = 바닥 -> 버리다
print(result01)

result02 = math.ceil(interest) # ceil = 천장 ->올리다
print(result02)
```

```python
#indexing 

colors = ["red", "green", "blue"]

# 1. index()로 위치를 먼저 알아내고
where_is_blue = colors.index("blue")  # 결과: 2

# 2. indexing으로 그 위치의 값을 확인 (또는 수정)
print(colors[where_is_blue])          # 결과: "blue"
```
