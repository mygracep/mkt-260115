```python
# url 값 통째로 찾아오는 방법 
import subprocess
import xml.etree.ElementTree as ET

url = "https://apis.data.go.kr/B551011/GoCamping/basedList?numOfRows=50&pageNo=1&MobileOS=ETC&MobileApp=camping&serviceKey=5b4e36add69e591ff769570c3ea884b7ab691337c5fdf7b7f71fa58c62368a01"

result = subprocess.run(["curl", "-k", url], capture_output=True, encoding = "utf-8") 
root = ET.fromstring(result.stdout)
print(root.findall(".//item")) # . : 최상위   // : 밑에 
```
