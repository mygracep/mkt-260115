```python
import requests
from bs4 import BeautifulSoup

url = "https://davelee-fun.github.io/"
headers = {"User-Agent" : "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/144.0.0.0 Safari/537.36 Edg/144.0.0.0"
}

response = requests.get(url, headers=headers)

if response.status_code!= 200:
    print("페이지 없음")
else :
    # 8회 반복 0~7까지 반복 
    for page_num in range(8) :
        if page_num == 0 : 
            response = requests.get(url, headers=headers)
        else : 
            response = requests.get(f"https://davelee-fun.github.io/page{str(page_num+1)}", headers=headers)
        
    
        soup = BeautifulSoup(response.content, "html.parser") 
        data = soup.select(".card-text")
        for item in data :
            print(item.get_text("",strip=True))


```

```python
import openpyxl 

def write_excel_template(filename, sheetname, listdata) :
    excel_file = openpyxl.Workbook()
    excel_sheet = excel_file.active
    excel_sheet.column_dimensions["A"].width = 100

    if sheetname !="" :
        excel_sheet.title= sheetname

    for item in listdata : 
        excel_sheet.append(item)
   
    excel_file.save(filename)
    excel_file.close()
```

```python
import requests
from bs4 import BeautifulSoup

url = "https://davelee-fun.github.io/"
headers = {"User-Agent" : "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/144.0.0.0 Safari/537.36 Edg/144.0.0.0"
}

response = requests.get(url, headers=headers)
product_lists = list() # or = [] 


if response.status_code != 200:
    print("페이지 없음")
    
else :
    # 8회 반복 0~7까지 반복 
    for page_num in range(8) :
        if page_num == 0 : 
            response = requests.get("https://davelee-fun.github.io", headers=headers)
        else : 
            response = requests.get(f"https://davelee-fun.github.io/page{str(page_num + 1)}", headers=headers)
        
    
        soup = BeautifulSoup(response.content, "html.parser") 
        data = soup.select(".card")
        for item in data :
            product_name = item.select_one(".card-text")
            product_date = item.select_one(".post-date")
            product_info = [product_name.get_text("", strip=True), product_date.get_text("", strip=True)]
            product_lists.append(product_info)
```

```python
write_excel_template("tmp.xlsx","상품정보", product_lists)
```
