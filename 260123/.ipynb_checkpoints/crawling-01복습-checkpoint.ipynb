- HTTP(Hyper Text Trenter Protocol) 상태코드
> 1xx (정보): 요청을 받았으며 프로세스를 계속한다
> 2xx (성공): 요청을 성공적으로 받았으며 인식했고 수용하였다
> 3xx (리다이렉션): 요청 완료를 위해 추가 작업 조치가 필요하다
> 4xx (클라이언트 오류): 요청의 문법이 잘못되었거나 요청을 처리할 수 없다
> 5xx (서버 오류): 서버가 명백히 유효한 요청에 대해 충족을 실패했다

```python
import requests
from bs4 import BeautifulSoup

url = "https://crawlingtest.netlify.app/"
headers = {"User-Agent" : "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/144.0.0.0 Safari/537.36 Edg/144.0.0.0"
}

respons = requests.get(url, headers=headers)
# respons.encoding = "utf-8" 
# soup = BeautifulSoup(respons.text, "html.parser")

soup = BeautifulSoup(respons.content, "html.parser") # 최근 문법 

# select / find 사용법 확인!! 
# items = soup.find_all("tr")

items = soup.select("tr")
for item in items :
    coulmns = item.select("td")
    for coulmn in coulmns :
        # print(coulmn.text.strip())
        print(coulmn.get_text(" ", strip=True))
```

```python
# urlib + bs4  사용 

from urllib.request import urlopen
from bs4 import BeautifulSoup

url = "https://davelee-fun.github.io/"
response = urlopen(url)
html = response.read().decode("utf-8")
soup = BeautifulSoup(html, "html.parser")

featured = soup.select_one(".featured-posts")
items = featured.select(".card-text")

for item in items :
    print(item.get_text(" ", strip=True))

```

```python
# http 상태 코드 확인 
import requests
from bs4 import BeautifulSoup

url = "https://davelee-fun.github.io"
headers = {"User-Agent" : "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/144.0.0.0 Safari/537.36 Edg/144.0.0.0"
}

respons = requests.get(url, headers=headers)

if respons.status_code!= 200:
    print("페이지 없음")
else :
    soup = BeautifulSoup(respons.content, "html.parser") 
    data = soup.select(".card-text")
    for item in data :
        print(item.get_text("",strip=True))
```
