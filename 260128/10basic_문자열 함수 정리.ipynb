### 문자열 관련 파이썬 기초문법

- count() : 문자열에 있는 특정 문자 갯수 세기 (단어 or 문자)
- index() : 문자열에 있는 특정 문자 위치 파악 ->!!  반드시 있는 값만 찾아올 수 있음 아니면 바로 error!!!
- find() : 문자열에 있는 특정 문자 위치 파악 but 애러발생 x  없을시 ->  "-1 "반환
- join() : () 안에 있는 문자열 사이사이에 . 앞에 있는 문자 삽입
- strip() : 문자열의 앞,뒤 공백 제거 (인자값 없을시)
 > rsstrip() 우측 공백 제거 // lstrip() 좌측 공백 제거
 > () 인자값 삭제 가능 but !!  중간값은 못건들임, 양 끝 값만 삭제 가능 
- upper() / lower(): 대,소문자로 변환 
- split() : 문자열 쪼갤때 -> 리스트를 반환 (인자값 없을 시 공백 기준으로 나눔)
- replace() :
