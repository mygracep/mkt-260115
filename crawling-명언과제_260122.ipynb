```python

url = "https://quotes.toscrape.com"
headers = {
    "Uuser-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/144.0.0.0 Safari/537.36 Edg/144.0.0.0"
}

import requests 
from bs4 import BeautifulSoup 

respons = requests.get(url, headers=headers)
respons.encoding = "utf-8" 

soup = BeautifulSoup (respons.text, "html.parser")

data_list = []
for item in soup.select(".quote") : 
    data_list.append({
         "명언" : item.select_one(".text").text.strip(),
         "저자" : item.select_one(".author").text.strip(),
         "테그" : item.select_one(".tags").get_text(" ", strip=True).split(":")[1].replace(" ", " -")

    })
print(data_list)


# word = [item.select(".text").text.scrip() for item in soup.select(".col-md-8")]
# print(word)
```
