# 데이터 자료구조 1 - list 

- 문자열, 정수, 실수, 논리형 -> (변수1 : 값1)
- 변수 : 여러값 그룹화
   - 공통분모를 가지고 있는 값들의 모임 -> 특성분류 가능
   - 컴퓨터 메모리 공강 효율적 사용 가능
   - 대부분의 크롤링 결과값 -> 리스트 형태의 자료구조 값이 결정

- 변수명 = [값, 값, 값] *대괄호 사용

- 리스트 & 문자열 : 유사성 띔. 비슷한 문법 구조
  - 공통속성 지님 /  _iter_ : iterable : 반복할 수 있는 / index
    
- 값을 추가하는 방법 
  - append() : 리스트 마지막(뒤)에 값을 추가
  - 리스트 + 리스트 연산
  - insert(추가하고자 하는 인덱스, 넣고자 하는 값) -> 원하는 위치에 원하는 값 추가
  - extend() : 여러개의 값을 한번에 넣고자 할때

- 값을 제거하는 방법
  - pop() : 리스트에서 값을 제거 (뒤에서부터) / pop(인덱스/위치)-> 인덱스 값 지움
  - remove() : 리스트에서 특정값을 직접 선택하여 지움
  - del 리스트명[인덱스] : 해당 인덱스 값 제거
  - 여러개 제거?? 

- slicing 기능 문자열과 동일 
- 리스트 자료구조 안에 입력될 값의 타입이 상이해도 가능
