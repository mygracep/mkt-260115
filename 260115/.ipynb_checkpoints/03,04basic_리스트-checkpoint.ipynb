# 데이터 자료구조 1 - list 

- 문자열, 정수, 실수, 논리형 -> (변수1 : 값1)
- 변수 : 여러값 그룹화
   - 공통분모를 가지고 있는 값들의 모임 -> 특성분류 가능
   - 컴퓨터 메모리 공강 효율적 사용 가능
   - 대부분의 크롤링 결과값 -> 리스트 형태의 자료구조 값이 결정

- 변수명 = [값, 값, 값] *대괄호 사용

- 리스트 & 문자열 : 유사성 띔. 비슷한 문법 구조
  - 공통속성 지님 /  _iter_ : iterable : 반복할 수 있는 / index
    
- 값을 추가하는 방법 
  - append() : 리스트 마지막(뒤)에 값을 추가 
  - extend() : 여러개의 값을 한번에 넣고자 할때 (맨 끝에 추가)
  - insert(index/위치, 넣고자 하는 값) -> 원하는 위치에 원하는 값 추가
  - 리스트 + 리스트 연산 = 제 3의 새로운 리스트 생성 
    수식: location = ["서울", "경기"] location += ["인천", "부산"] # 결과: ['서울', '경기', '인천', '부산']

 

- 값을 제거하는 방법
  - pop() : pop(인덱스/위치)-> 인덱스 값 지움  리스트 뒤에서부터 값을 제거 
  - remove() : 리스트에서 특정값을 직접 선택하여 지움
  - 다수의 값 삭제 : del 리스트명[인덱스] : 해당 인덱스 값 제거/ slicing과 동일 
    

- slicing 기능 문자열과 동일 
- 리스트 자료구조 안에 입력될 값의 타입이 상이해도 가능

- 리스트 값 정렬 (오름차순/ 내림차순)
  - sort : 오름차순 정렬 (원본 데이터 변경)
  - reverse : 내림차순 정렬 (원본 데이터 변경)

- split() : 쪼개다/ 문자열을 리스트로 변경해주는 기능 (원본 편집/변경 x )

```python
location = ["서울", "경기", "인천"]

# print(type(location))
# print(location)
print(location[0])
print(location[1])
print(location[2])
print(location[-1])
```

```python
location = ["서울", "경기", "인천"]
location.append("부산")
location.extend("인천")
location.extend(["인천"])
print(location);
```

```python
# 값을 추가하는 방법 
a = [3, 4, 5]
b = [1, 2]

result = b + a
result.insert(0, 0)
result.insert(len(result), 0) # 맨 마지막에 넣고 싶을때  
print(result)
```

```python
# 여러개 값 추가 

a = [1, 2, 3]

a.append([4, 5]) # [4,5] 하나의 값으로 추가됨 
print(a[3])

a.extend([4, 5])
print(a)
print(a[5])
```
