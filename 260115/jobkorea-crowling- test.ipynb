```python
# selenium

from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By 
from webdriver_manager.chrome import ChromeDriverManager
# import urllib.parse -> 언어(한국어) 인코딩 되는 값 미리 설정 
from urllib import parse
import time # 너무 빨리 작업하지 않도록 시간 벌이용 
import pandas as pd # as= 축약하겠다 


service = Service(ChromeDriverManager().install())
options = Options()
# options.add_argument("--headliss") # 가상의 브라우저를 우리 눈에 보이지 않게 설정 
# options.add_argument("--disable-gpu") 
options.add_argument("--disable-dev-shm-usage") 
options.add_argument("--window-size=1920x1080") 
options.add_argument("--start-maxized") 
options.add_argument("--user-agent= Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/143.0.0.0 Safari/537.36") 
options.add_argument("--lang=ko_KR") 

driver = webdriver.Chrome(service=service, options=options)

query = "퍼포먼스마케터"
encoded_query = parse.quote(query)

max_pages = 6

results = []


# range() = 범위(반복횟수) nut 0부터 시작 
for page in range(1, max_pages + 1) : 
    url = f"https://www.jobkorea.co.kr/Search?stext={encoded_query}&Page_No={page}" # no=3 페이지 숫자 지우고 설정한 page 값으로 넣음 
    driver.get(url)
    time.sleep(2)
    
    job_posts = driver.find_elements(By.CSS_SELECTOR, "div.styles_mb_space2__dk46ts4t")

    for post in job_posts : 
        try : 
            title_elem =  post.find_element(By.CSS_SELECTOR, "a.Flex_display_flex__i0l0hl2.Flex_gap_space6__i0l0hl1g.Flex_align_center__i0l0hl8.styles_mb_space2__dk46ts4t")
            print(title_elem.text.strip()) 
            link = title_elem.get_attribute("href")
            results.append( {
              "공고제목": title,
              "지원링크": link
            })
        except : 
            continue 


    
driver.quit()  

df = pd.DataFrame(results)

df.to_excel("잡코라아_퍼포먼스마케터.xlsx", index=False)
print("잡코라아_퍼포먼스마케터.xlsx 저장완료")


# f 스트링 문법: f 스트링을 안쓰면 문자/변수를 계속 구분해서 연산기호를 넣어줘야함 
# f 스트링 사용시 변수 {} 중괄호 처리 
```
