# 파이썬 기초문법
- 출력 : print() > 내장함수 = built in function / 내장 : 애초에 파이선 문법 탄생될 때부터 존재
- 입력 : input("당신의 이름은?") > 사용자로부터 어떤 값을 입력받을 수 있음 / placeholder = 입력해야할 값의 정보
   - 해당함수를 통해서 수집한 값은 무조건 "문자열"이 된다는 사실을 기억해라!
- 연산 :
   - 사칙연산 : + @ → + - * / %(나누기 연산에 대한 나머지) ** (제곱) ※ '+' : 문자간의 연결연산자로도 사용가능!
   - 비교연산 : <, >, <=, >=, ==(같다), !=(같지 않다) ※ (좌항 = 우항, 우항의 값을 좌항에 삽입&적용)
- 변수 :
   - variable : 변할 수 있는 값은 넣는 공간 = 그릇
   - 변수명 작업 : 직관, 명시 → 누가봐도 한눈에 알아볼 수 있는 변수명 / 숫자가 먼저올 수 없음 (중간에 오는 것은 가능)
   - 팁 : 스네이크 표기법 (뱀 → case_study / 카멜 표기법 → caseStudy)
    
- 자료형 : 변수 안에 담을 수 있는 값의 형태 → 문자열(String)"" VS 문자 (Character)'' / 정수(Int) / 실수(Float) / 논리형(Boolean)
   - 형변화 : 자동형변환 → input() / 명시적형변환 → 수동적으로 형태를 변환
   - 자료형확인 : type()
   - int() : 정수형 변환
   - float() : 실수형 변환
   - str() : 문자열 변환
   - bool() : 논리형 변환
   - list() : 리스트 자료구조 변환
   - dict() : 딕셔너리 자료구조 변환
   - tuple() : 튜플 자료구조 변환
   - set() : 세트(집합) 자료구조 변환
     
- 문자열 : 독립적인 함수 혹은 속성들이 많이 존재
   - count() : 특정 문자열에서 어떤 단어가 몇 번 사용되었는지를 집계하고자 할 때 (대, 소문자 구분)
   - len() : length = 길이
   - find() : 해당 문자열에서 인자값으로 전달된 문자가 몇번쨰 인덱스(위치)에 위치하는지 찾고자 할 때
   - replace() : 특정 문자열을 대체하고자 할때
   - index() : 문자열 안에 각각의 문자는 고유한 숫자를 가지고 있음 -> 문자열 왼쪽부터 시작 / 무조건 0부터 시작
   - slicing() : 문자열 안에서 특정 구간을 잘라내서 가져오고 싶을 때
   - strip() : 특정 문자열 앞, 뒤쪽에 여백 혹은 여백으로 인지할 수 있는 escape sequence(\n,\t 공백) 문자열을 제고하고자 할때
   - formet() : 특정 문자열을 원하는 위치에 출력하고자 할때 사용할 수 있는 메서드(method) 함수

- 형식지정자 : 특정 타입의 값을 출력하고자 할때 어떤 형태로 값을 찾아와서 훌력할 것인가 결정 
   - %s - string 문자열
   - %c - character
   - %d - int 정수
   - %f - float 실수

```python
#find

text = "banana"

index = text.find("a")
next_index = text.find("a", index + 1) # 1번째, 그 다음번째( 2번째) 인덱스 찾기 
next_index = text.find("a", 3 + 1) # 3=3번째 위치한 문자 /  +1= 그 다음번째 인덱스 
print(index, next_index)
```

```python
#반복문 사용 

text = "banana"
target = "a"

idx = text.find(target) # 존재하지 않는 문자를 찾을 경우 = -1 값 

# 예외조항 처리 : 어떤 명령문을 실행시키고자 할때, 만약 ~가 아니라면, 해라!! 

while idx != -1 : 
    print(idx)
    idx = text.find(target, idx + 1)
```

```python
text = "python"
print(text) # 문자 통채로 찾아오기 
print(text[5]) # ~번째 인덱스 찾아옴  
# 어던 문자열이던, 마지막번째 문자의 인덱스 넘버는 전체 문자열 개수 -1 

print(len(text)) # 문자 개수
print(len(text)-1) # 전체 -1 개수
print(text[len(text)-1]) # 전체 개수 -1 의 문자 출력
print(text[-1]) # 오늘쪽, 가장 끝에 문자를 찾는건 -1부터 시작 
```

```python
#slicing

text = "python"
print(text)
print(text[2])

print(text[2:4]) # 슬라이싱: 찾아오고자 하는 문자열 [start:finish+1]
print(text[2:]) # 슬라이싱에서 마지말번째 인덱스의 생략 = 끝까지 
print(text[2:len(text)]) #->해당 구문 많이 사용 
#=> 마지막까지 찾아올 수 있는건 이 3가지 문구 

print(text[2:-1]) #마지막 바로 전 문자까지 찾기 가능 
```

```python
# strip 

some_str = "  computer  "
print(some_str)
print(some_str.strip()) # 보통 괄호안에 값을 넣지 않음 

#특정 문자를 날리고 싶을때 
some_str01 = "...computer,,,"
print(some_str01.strip(","))
print(some_str01.strip("."))
print(some_str01.strip(",."))
```

```python
#farmet 
# 객체 - 메서드 함수 
# 형식: 문자열 형식의 객체(반드시 필요). formet()

name = "I have a {}, I have an {}."
print(name)

print(name.format("pen", "apple")) # 포멧 괄호 인자값은 순차적으로 중괄호에 들어감 
print("I have a {1}, I have an {0}".format("pen", "apple")) # 포멧 괄호 인자값은 순차적으로 중괄호에 들어감 
```

```python
interest = 0.0891
print(format(interest, ".2f")) # 독립함수 (문자, 숫자 x) -> 소수점 위치 일치 

#파이썬 내장함수 가운데 소수점을 올리거나 내리는 독립 기능 x
#외부 라이브러리 모듈 찾아서 사용 
#수학갹체 라이브러리 모듈 

import math 
result01 = math.floor(interest) # floor = 바닥 -> 버리다
print(result01)

result02 = math.ceil(interest) # ceil = 천장 ->올리다
print(result02)
```

```python
print("I have a %s, I have an %s." % ("pen", "apple")) 
print("I have a %f, I have an %d." % (3.14, 7)) 
print("I have a %f, I have an %c." % (3.14, 'A')) #엄격한 프로그래밍 언어-> "" 문자열, '' 문자
print("I have a %f, I have an %c." % (3.14, 97)) # 문자는 정수로 치환됨 
print("I have a %f, I have an %s." % (3.14, "Ab"))  # 문자 2개 이상은 -> 문자열 
```
