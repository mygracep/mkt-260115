# 반복문

## for, whlie 
* 반복문/ 조건문을 사용할때 실행문은 항상 들여쓰기!!

- for : 구체적인 반복 횟수 정해져있을때 
- whlie : 구체적인 반복횟수가 정해져 읺지 않을 때 / 무한루프 -> 조건 충족시 종료



- for 변수 in (반복할 대상/개체 = 속성 "iter" = 리스트/문자열)
   실행문  0 ~  

- range: 반복될 것이 정확하게 명시되지 않을때, 범위를 잡아서 작동
       실행문 0 ~   
 for 변수 in range (인수= 어떤 숫자 시작, 인수= 해당 숫자에서 -1한 값만큼) :  
       실행문 0 ~

- while 조건식:
       실행문 


```python
for index in range(10) : 
    print(index)
```

```python
# 누산기
sum = 0  # 아무런 값에 영향을 미치지 않는 값으로 0 선언 

for index in range(1, 11) : 
    # sum = sum + index  # 누산기
    sum += index  # 복합대입 연산자
    
print(sum)
```

# 복합 대입 연산자 
sum = sum + index 
->  sum += index  

a = a + b  ->  a += b 
a = a - b  ->  a -= b 
a = a * b  ->  a *= b 
a = a / b  ->  a /= b 
a = a % b  ->  a %= b 
a = a ** b  ->  a **= b 

```python
name = ""
while name != "David" :
    name = input("whats your name?")
    if name != "David" :
     print(f"Hi~ {name}, you konw David?")
    else : 
        print(f"Hi~{name}")
```

```python
num = int(input("몇 단의 구구단을 출력?"))

for digit in range(1,10) : 
    print(F"{num} X {digit} = {num * digit}")
```

```python
# 1~30까지 (정수) 숫자 중 3의 배수만 출력 

num = 0 

while num < 30 : # 0 ~ 29번째 까지 반복해라 
    num+= 1 
    if num % 3 == 0 :
        print(num)
```
