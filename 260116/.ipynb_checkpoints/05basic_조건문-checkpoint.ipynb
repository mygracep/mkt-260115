# 조건문 : 값을 출력하기 위해서, A / B에 따라 다른 값 출력 

1 if 조건문 : A라면 ~ / B라면 ~ (단일조건) 
2 if, else문 : 만약 A가 아니라면 (양자택일)  
3 if elif else문 : A라면 ~ / B라면 ~ /C라면~  (다중조건)
  * else문 : 무조건 if, elif 보다 먼저 나올 수 없다 / 나머지, 그밖에 
  * else문 : 비필수 조건
  * 조건문 : 구조 -> (조건식 + 실행문)

- 조건문 > 조건식 > 비교연산자 (~보다 ~가 크다) / 논리연산자( ~보다 크다고 한다면,)

```python
age1 = input("민수의 나이는?")
age_digit1 = int(age1)

age2 = input("강인의 나이는?")
age_digit2 = int(age2)

if age_digit1 >= 19 and age_digit2 >= 19 : 
    print("두 사람 모두 성인입니다")
```

```python
age = int(input("당신의 나이는?"))

if age >= 19 :
    print("당신은 성인입니다.")

elif age >= 13 and age < 19 : 
    print("당신은 청소년입니다.")

elif age >= 5 and age < 13 : 
    print("당신은 청소년입니다.")

else : 
    print("당신은 아동")
```

```python
# input("현재 가지고 있는 돈은 얼마?") -> 
cash = int(input("현재 가지고 있는 돈은 얼마?"))

if cash > 100000 :
    print("Go to find resturant")
else :
    if cash > 50000 :
       print("Go to korean resturant")
    else : 
     print("go to home")
```
