```python
numbers = []
for i in range(3) :
    number = int(input(f"{i + 1}번째 숫자 : ")) #컴퓨터가 인식하는 숫자여야 하니까 +1 
    numbers.append(number)


nb1, nb2, nb3 = numbers
# 구조분해할당 문법: 해당값이 인데스와 매칭 /같은자리끼리 짝지어짐 
# ㄴ> 리스트와 같이 여러개 변수값을 가지고 있는 자료구조의 값을 하나씩 추출 > 독립변수로 담을 수 있는 문법 

if nb1 <= nb2 and nb1 <= nb3 :
     print(nb1)
elif nb2 <= nb1 and nb2 <= nb3 :
     print(nb2)
elif nb3<= nb1 and nb3 <=nb2 :
     print(nb3)
```

```python
numbers = []
for i in range(3) : # range : index가 작동하는 범위/  반복 횟수 or 반복구간설정 /  ~이상 ~ 미만 
    number = int(input(f"{i + 1}번째 숫자 : ")) #컴퓨터가 인식하는 숫자여야 하니까 +1 
# range (반복구간 설정)  ~이상 ~ 미만
for i in range(1,4) :
    number = int(input(f"{i}번째 숫자 : "))    
    numbers.append(number)

print(min(numbers))
```

```python
# 리스트 컴프리핸션 - list comrehension

numbers = [int(input(f"{i +1}번째 숫자: ")) for i in range(3) ]

print(min(numbers))
```

```python
# 주민벟호 받아서 성별 구분해보기 

pn = input("주민번호는?")
gender = pn.split("-")[1][0]

if gender == "1" or gender == "3" : # gender 가 문자이기 때문에 2 사용 불가. 둘중에 하나 바꿔야 함 
     print("남자 입니다.")
elif gender == "2" or gender == "4" : 
    print("여자")

else : 
    print("잘못된 주민번호")
```
